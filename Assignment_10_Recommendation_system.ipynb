```python
books_df = book1.pivot_table(index ='User.ID' ,columns ='Book.Title' , values ='Book.Rating').reset_index(drop =True)
```

```python
books_df 
```

```python
# check userid
df_user = book1['User.ID'].unique()
df_user
```

```python
df_userid = pd.DataFrame(df_user ,columns=['userid'])
```

```python
# Sort userid in ascending order
df_userid1 = df_userid.sort_values('userid' ,ascending = True)
df_userid1
```

```python
books_df.index = df_userid1.userid.unique()
```

```python
books_df 
```

```python
#Impute those NaN with 0 values
```

```python
books_df.fillna(0 , inplace = True)
```

```python
books_df
```

# Calculating Cosine Similarity between Users

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
# calculate distance
pair_dist = pairwise_distances(books_df.values , metric = 'cosine') #.values = it taken rating values
pair_dist
```

```python
# evaluate similarities
user_sim = 1 - pair_dist
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
# Set the index and column names to user ids
user_sim_df.index = book1['User.ID'].unique()
user_sim_df.columns = book1['User.ID'].unique()
user_sim_df
```

```python
# Nullifying diagonal values

np.fill_diagonal(user_sim,0)
user_sim_df
```
